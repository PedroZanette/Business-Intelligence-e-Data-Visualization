```python
import pandas as pd
import numpy as np

df = pd.read_csv("EmendasParlamentares.csv", sep=";", encoding="ISO-8859-1", low_memory=False)

money_cols = [
    "Valor Empenhado", "Valor Liquidado", "Valor Pago",
    "Valor Restos A Pagar Inscritos", "Valor Restos A Pagar Cancelados",
    "Valor Restos A Pagar Pagos"
]

for c in money_cols:
    df[c] = (df[c].astype(str)
                .str.replace(".", "", regex=False)   # tira separador de milhar (se tiver)
                .str.replace(",", ".", regex=False)) # troca vírgula por ponto
    df[c] = pd.to_numeric(df[c], errors="coerce")

df["Ano da Emenda"] = pd.to_numeric(df["Ano da Emenda"], errors="coerce")
```

```python
# Visualização inicial do dataset após tratamento dos dados
import pandas as pd

pd.options.display.float_format = '{:,.2f}'.format

print("Primeiras 3 linhas do DataFrame:")
display(df.head(3))

print("\nEstatísticas descritivas das colunas monetárias:")
display(df[money_cols].describe())
```

### 1. Qual o valor total pago em emendas por ano? (Utiliza: groupby e sum na coluna Ano da Emenda e Valor Pago).

```python
# Agrupei os dados por "Ano da Emenda" e utilizei a função sum()
# para calcular o valor total efetivamente pago em cada ano
# A soma foi escolhida pois a pergunta solicita o total acumulado
total_pago_ano = df.groupby("Ano da Emenda")["Valor Pago"].sum().sort_index()
total_pago_ano
```

### 2. Qual é a média e o desvio padrão dos valores empenhados por região do Brasil? 


```python
# Agrupei por "Região" e utilizei agg(["mean","std"])
# para calcular a média e o desvio padrão dos valores empenhados
# A média mostra o valor típico e o desvio padrão indica a dispersão.
stats_regiao = df.groupby("Região")["Valor Empenhado"].agg(["mean", "std", "count"]).sort_values("mean", ascending=False)
stats_regiao
```

### 3. Quais são os 10 autores que mais destinaram recursos (valor empenhado) na história do dataset? 


```python
# Agrupei por "Nome do Autor da Emenda" e somei o "Valor Empenhado"
# Em seguida ordenei em ordem decrescente e selecionei os 10 maiores valores.
# O valor empenhado foi usado por representar a intenção orçamentária

top10_autores = (df.groupby("Nome do Autor da Emenda")["Valor Empenhado"]
                   .sum()
                   .sort_values(ascending=False)
                   .head(10))
top10_autores
```

### 4. Quantas emendas foram destinadas para o estado de Santa Catarina?


```python
# Filtrei os registros onde a coluna "UF" corresponde a Santa Catarina
# e utilizei a contagem de linhas para determinar o número total de emendas destinadas ao estado.

qtde_sc = int((df["UF"].str.upper() == "SANTA CATARINA").sum())

print(f"Quantidade de emendas destinadas a Santa Catarina: {qtde_sc}")
```

### 5. Destas emendas, quais municípios se destacam no recebimento dos recursos por tipo?



```python
# Filtrei apenas o estado de Santa Catarina e agrupei por
# "Tipo de Emenda" e "Município", somando o "Valor Pago"
# Utilizei Valor Pago por representar a execução real do recurso.

sc = df[df["UF"].str.upper() == "SANTA CATARINA"].copy()

sc_clean = sc[~sc["Município"].isin(["Múltiplo", "Sem informação"])]

ranking = (sc_clean.groupby(["Tipo de Emenda", "Município"])["Valor Pago"]
             .sum()
             .reset_index()
             .sort_values(["Tipo de Emenda", "Valor Pago"], ascending=[True, False]))

top_por_tipo = ranking.groupby("Tipo de Emenda").head(5)
top_por_tipo
```

### 6. Existem emendas onde o valor liquidado é significativamente diferente do valor pago? 


```python
# Calculei a diferença entre "Valor Liquidado" e "Valor Pago"
# para identificar possíveis discrepâncias na execução orçamentária
# Utilizei o valor absoluto para destacar diferenças significativas.

df["diff_liq_pago"] = df["Valor Liquidado"] - df["Valor Pago"]
df["abs_diff"] = df["diff_liq_pago"].abs()

# Top 10 maiores diferenças absolutas
df.sort_values("abs_diff", ascending=False)[
    ["Ano da Emenda","UF","Município","Nome do Autor da Emenda","Valor Liquidado","Valor Pago","diff_liq_pago"]
].head(10)
```

### 7. Qual o percentual de recursos que ficaram como "Restos a Pagar Cancelados" em relação ao total empenhado por ano? 


```python
# Agrupei por ano e somei o "Valor Restos A Pagar Cancelados"
# e o "Valor Empenhado". Em seguida calculei a proporção
# entre cancelados e empenhados para obter o percentual anual

agr = df.groupby("Ano da Emenda")[["Valor Restos A Pagar Cancelados","Valor Empenhado"]].sum()
agr["pct_cancelado"] = np.where(agr["Valor Empenhado"] > 0,
                                agr["Valor Restos A Pagar Cancelados"] / agr["Valor Empenhado"],
                                np.nan)
agr[["pct_cancelado"]]
```

### 8. Qual é a Subfunção mais comum para cada Região do país? 



```python
# Realizei a contagem de ocorrências por "Região" e "Nome Subfunção"
# e selecionei a subfunção com maior frequência em cada região
# A escolha foi baseada na maior quantidade de registros

cont = df.groupby(["Região","Nome Subfunção"]).size().reset_index(name="n")
idx = cont.groupby("Região")["n"].idxmax()
mais_comum = cont.loc[idx].sort_values("Região")
mais_comum
```

### 9. Existem linhas com Cdigo Municpio IBGE ausente? Como isso afeta a análise por localidade? 


```python
# Verifiquei valores ausentes ou inválidos na coluna
# "Código Município IBGE" utilizando isna() e filtros adicionais
# Isso é importante pois dados ausentes impactam análises por localidade.

col = "Código Município IBGE"
ausente = df[col].isna() | df[col].astype(str).str.strip().isin(["", "S/I", "Sem informação"])
total_ausentes = int(ausente.sum())
percentual_ausentes = round(ausente.mean() * 100, 2)

print(f"Total de registros com Código Município IBGE ausente: {total_ausentes}")
print(f"Percentual de registros ausentes: {percentual_ausentes}%")
```

### 10. O campo Município possui nomes duplicados com grafias diferentes (ex: acentuação)? 

```python
# Normalizei os nomes dos municípios removendo acentos e padronizando para maiúsculas.
# Em seguida verifiquei se múltiplas grafias representavam o mesmo município.
# Essa etapa melhora a consistência da análise.


from unidecode import unidecode

tmp = df[["Município"]].dropna().copy()
tmp["norm"] = tmp["Município"].astype(str).str.strip().map(lambda x: unidecode(x).upper())

variantes = tmp.groupby("norm")["Município"].nunique().sort_values(ascending=False)
variantes[variantes > 1].head(20)
```
